```python
#Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
data = pd.read_csv('/content/Bank Customer Churn Prediction.csv')
```

**Univariate analysis** gives you *insights* into individual variables, while **bivariate analysis** explores *relationship*s between two variables.

1**. Univariate Analysis**


**a.Numeric Variables**

```python
# Histogram for numeric variables
data.hist(figsize=(12, 10))
plt.show()

# Boxplot for numeric variables
sns.boxplot(data=data, orient="h", palette="Set2")
plt.show()
```

**b.Categorical Variables**

```python
# Countplot for categorical variables
sns.countplot(x='credit_score', data=data, palette='viridis')
plt.show()
```

**c. Target Variable Analysis**

```python
#Target Variable Analysis
# For classification problems
sns.countplot(x='churn', data=data, palette='Set1')
plt.show()

# For regression problems
sns.histplot(data['churn'], kde=True, color='blue')
plt.show()
```

**Bivariate Analysis**

```python
#Numeric vs. Numeric
# Scatter plot of 'age' vs. 'balance'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='balance', data=data, hue='churn', palette='viridis')
plt.title('Scatter Plot of Age vs. Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
#Categorical vs. Numeric
# Box plot of 'credit_card' vs. 'balance'
plt.figure(figsize=(8, 6))
sns.boxplot(x='credit_score', y='balance', data=data)
plt.title('Box Plot of credit_card vs. Balance')
plt.xlabel('credit_card')
plt.ylabel('Balance')
plt.show()

```

```python
# Categorical vs. Categorical
# Cross-tabulation of 'country' and 'churn'
country_churn_cross = pd.crosstab(data['country'], data['churn'])
print("\nCross-tabulation of Country and Churn:")
print(country_churn_cross)

# Heatmap of the cross-tabulation
plt.figure(figsize=(8, 6))
sns.heatmap(country_churn_cross, annot=True, cmap='Blues', fmt='d')
plt.title('Heatmap of Country vs. Churn')
plt.xlabel('Churn')
plt.ylabel('Country')
plt.show()
```

```python
#Correlation Matrix
# Compute the correlation matrix for numeric variables
correlation_matrix = data.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Pairwise Plots
# Pairwise scatter plots for numeric variables
sns.pairplot(data[['credit_score', 'age', 'balance', 'estimated_salary', 'churn']], hue='churn', palette='Set1')
plt.title('Pairwise Plots of Numeric Variables')
plt.show()
```

**Multivariate Analysis** involves the simultaneous analysis of multiple variables to understand the relationships among them. Unlike univariate analysis (which focuses on a single variable) and bivariate analysis (which examines the relationship between two variables), multivariate analysis deals with three or more variables.

```python
# Multivariate Analysis - Pairwise Plots with Hue
# Pairwise scatter plots for numeric variables with 'churn' as the hue
sns.pairplot(data[['credit_score', 'age', 'balance', 'estimated_salary', 'churn']], hue='churn', palette='Set1')
plt.title('Pairwise Plots with Churn as Hue')
plt.show()
```

```python
# Multivariate Analysis - Pairwise Plots with Multiple Variables
# Pairwise scatter plots for selected numeric variables
sns.pairplot(data[['age', 'balance', 'estimated_salary', 'products_number', 'churn']], hue='churn', palette='Set2')
plt.title('Pairwise Plots with Multiple Variables')
plt.show()
```
