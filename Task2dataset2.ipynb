```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Electric_Vehicle_Population_Data.csv')
```

```python
print(data.head())
```

Univariate Analysis

```python
def univariate_analysis(variable):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[variable], kde=True)
    plt.title(f'Univariate Analysis of {variable}')
    plt.show()
```

```python
univariate_analysis('Electric Utility')
```

Bivariate Analysis

```python
def bivariate_analysis(x_variable, y_variable):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data[x_variable], y=data[y_variable])
    plt.title(f'Bivariate Analysis: {x_variable} vs {y_variable}')
    plt.show()
```

```python
bivariate_analysis('Electric Vehicle Type', 'Electric Range')
```

Multivariate Analysis

```python
def multivariate_analysis(variables):
    plt.figure(figsize=(12, 8))
    sns.pairplot(data[variables])
    plt.title('Multivariate Analysis')
    plt.show()
```

```python
multivariate_analysis(['Electric Utility', 'Electric Range', 'Base MSRP'])
```
